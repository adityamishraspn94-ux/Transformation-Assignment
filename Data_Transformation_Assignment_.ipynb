# **Q1. Handle Missing Values (Basic Cleaning)**

## **Task:**

Fill missing values using:

* Age → Median
* Salary → Mean
* Hours_Worked_Per_Week → Median
* Performance_Score → Mean

## C**ode**

```python id="3v8w5h"
import pandas as pd
import numpy as np

df = pd.read_csv("employee_productivity_dataset.csv")

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)

df['Salary'].fillna(df['Salary'].mean(), inplace=True)

df['Hours_Worked_Per_Week'].fillna(
    df['Hours_Worked_Per_Week'].median(),
    inplace=True
)

df['Performance_Score'].fillna(
    df['Performance_Score'].mean(),
    inplace=True
)

print(df.head())
```
## **Explanation:**

* Median is used for Age and Hours_Worked_Per_Week because they may contain outliers.
* Mean is used for Salary and Performance_Score because they are continuous numerical variables.

---

# **Q2. Label Encoding**

## **Task:**

Convert:

* Gender
* Department

into numerical form using Label Encoding.
## **Code**

```python id="r6k0q2"
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])

df['Department'] = le.fit_transform(df['Department'])

print(df[['Gender', 'Department']].head())
```

## **Explanation:**

Label Encoding converts categorical text values into numeric values.

Example:

* Male → 1
* Female → 0

## **Explanation:**

* ColumnTransformer applies different preprocessing steps on different columns.
* Pipeline automates preprocessing workflow.
